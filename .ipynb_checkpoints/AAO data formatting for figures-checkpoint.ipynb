```python
# turning into colors--read out observation functions -- this doesn't seem to work yet.
obs_file = fileDir + '/observationfunctions.csv'
obs_raw = pd.read_csv(obs_file, index_col = 'wl')
# plt.plot(obs_raw)
fx = interp1d(obs_raw.index.values, obs_raw['r'])
fy = interp1d(obs_raw.index.values, obs_raw['g'])
fz = interp1d(obs_raw.index.values, obs_raw['b'])

D65 = pd.read_csv(fileDir + '/d65.csv', index_col ='wl')
fD65 = interp1d(D65.index.values, D65['d65'])
# plt.plot(D65)
# plt.plot(D65.index.values, fD65(D65.index.values))
```

```python
path = fileDir + '/AAO data files reformatted/specular' # alternates: '/AAO files unformatted/specular' and '/AAO files unformatted/NA'
files = os.listdir(path)
files.sort()
print([x for x in files if 'csv' in x])
datafiles = [x for x in files if 'csv' in x]
```

```python
datatypes = ['0','45', '60', '75']
hexcolors = pd.DataFrame(index = datatypes)
labcolors = pd.DataFrame(index = datatypes)
for file in datafiles:
    #file = datafiles[0]
    data = pd.read_csv(path + '/' + file)

    #datatypes
    step = WVL[1] - WVL[0]

    colors = []
    labc = []
    for d in datatypes:
        if d in data.columns:
            #print(d)

            X = 0
            Y = 0
            Z = 0
            k = 0
            for i in range(len(WVL)):
                X += data[d][i] * fD65(WVL[i]) * fx(WVL[i]) * step 
                Y += data[d][i] * fD65(WVL[i]) * fy(WVL[i]) * step
                Z += data[d][i] * fD65(WVL[i]) * fz(WVL[i]) * step
                k += fD65(WVL[i]) * fy(WVL[i]) * step

            k = 100/k
            X = X*k/100
            Y = Y*k/100
            Z = Z*k/100

            # return as a color object
            colors += [spectra.xyz(X,Y,Z).hexcode]
            labc += [spectra.xyz(X,Y,Z).to('lab').values]
            #print(color.hexcode)
        else:
            print(d, ' not in dataset. adding placeholder.')
            colors += ['N/A']
            labc += ['N/A']
    #print(colors)
    hexcolors[file] = colors
    labcolors[file] = labc
    #print([c.hexcode for c in colors if type(c) != str])
```

```python
datatypes = ['.3','.45', '.6', '.75']
hexcolors = pd.DataFrame(index = datatypes)
labcolors = pd.DataFrame(index = datatypes)
xyycolors = pd.DataFrame(index = datatypes)
colordf = pd.DataFrame(index = datatypes)
hsvcolors = pd.DataFrame(index = datatypes)
for file in datafiles:
    #file = datafiles[0]
    data = pd.read_csv(path + '/' + file)

    #datatypes
    step = WVL[1] - WVL[0]
    
    XYZcolors = []
    colors = []
    labc = []
    xyy = []
    hsvlist = []
    for d in datatypes:
        if d in data.columns:
            #print(d)

            X = 0
            Y = 0
            Z = 0
            k = 0
            for i in range(len(WVL)):
                X += data[d][i] * fD65(WVL[i]) * fx(WVL[i]) * step 
                Y += data[d][i] * fD65(WVL[i]) * fy(WVL[i]) * step
                Z += data[d][i] * fD65(WVL[i]) * fz(WVL[i]) * step
                k += fD65(WVL[i]) * fy(WVL[i]) * step

            k = 100/k
            X = X*k/100
            Y = Y*k/100
            Z = Z*k/100

            # return as a color object
            XYZcolors +=[spectra.xyz(X,Y,Z).values]
            colors += [spectra.xyz(X,Y,Z).hexcode]
            labc += [spectra.xyz(X,Y,Z).to('lab').values]
            xyy += [np.array([X/(X+Y+Z),Y/(X+Y+Z)])]
            hsvlist += [spectra.xyz(X,Y,Z).to('hsv').values]
            
            #print(color.hexcode)
        else:
            print(d, ' not in dataset. adding placeholder.')
            colors += ['N/A']
            labc += ['N/A']
            
            print(xyy)
    #print(colors)
    colordf[file] = XYZcolors
    hexcolors[file] = colors
    labcolors[file] = labc
    xyycolors[file] = xyy
    hsvcolors[file] = hsvlist
    #print([c.hexcode for c in colors if type(c) != str])
```

```python
# now interpolate each series to get colors
D65 = pd.read_csv(fileDir + '/d65 from 300.csv', index_col ='wl')
fD65 = interp1d(D65.index.values, D65['D65'])

wvl = vase.index.values.astype(np.float)
step = wvl[1] - wvl[0]
X = 0
Y = 0
Z = 0
k = 0
for dtype in vase.columns:
    for WVL in vase.index.values:
        R = vase[dtype][WVL]
        X += R * fD65(WVL) * fx(WVL) * step
        Y += R * fD65(WVL) * fy(WVL) * step
        Z += R * fD65(WVL) * fz(WVL) * step
        k += fD65(WVL) * fy(WVL) * step

    k = 100/k
    X = X*k/100
    Y = Y*k/100
    Z = Z*k/100

    color = spectra.xyz(X,Y,Z)
    plt.figure(figsize = [3,3])
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xticks([])
    plt.yticks([])
    poly_x = [0,  1,  1, 0]
    poly_y = [0, 0, 1, 1]
    color_string = color.hexcode
    print(color_string)
    plt.fill(poly_x, poly_y, color_string, edgecolor=color_string)
```
